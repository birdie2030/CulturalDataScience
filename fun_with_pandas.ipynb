## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
```

```python
# making a dataframe

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
df_physical = rawdata[0]
df_physical['Media'] = "Physical"
df_physical
```

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
df_digital = rawdata[3]
df_digital['Media'] = "Digital"
df_digital
```

```python
# making a combined df

df_combined = pd.concat([df_physical,df_digital])
df_combined
```

```python
# Artist and single name for closest year to when I was born

birth = df_combined[df_combined['Released']==2008]
birth[['Artist', 'Single']]
```

```python
# Artist and single for year I turned 15
birth2 = df_combined[df_combined['Released']==2015]
birth2[['Artist', 'Single']]
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
# making a dataframe

df_solarsystem = pd.concat(rawdata[0:8])
df_solarsystem
```

```python
# splitting the date into three types
df_solarsystem[['Day', 'Month', 'Year']] = df_solarsystem['Launch date'].str.split(' ', expand = True)
print (df_solarsystem)
```

```python
# printing new dataframe with splitted dates
df_solarsystem
```

```python
# finding out how many launches there were for each year
most_launches = df_solarsystem['Year']
most_launches.value_counts()
```

```python
# importing statistics package
import statistics as stat
```

```python
# Finding the year with the most launches
stat.mode(most_launches)
```

```python
# Finding the most common month for launches
common_month = df_solarsystem['Month']
common_month.value_counts()
stat.mode(common_month)
```

```python
# printing a list that ranks the months from most launches to fewest launches

common_month.value_counts()
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
# making a dataframe
villaindata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")

df_villains = pd.concat(villaindata[3:12])
df_villains
```

```python
# Challenge 2. Defining the decades
decades = ['1930s', '1940s', '1950s','1960s','1970s','1980s','1990s','2000s','2010s']
```

```python
# Creating a dataframe
villains = pd.DataFrame()
```

```python
# Making a loop that looks through the tables and assigns a decade to them
for i,decade in enumerate(decades):
    table = villaindata[i+3]
    table['Decade'] = decade
    villains = pd.concat([villains, table], ignore_index=True)

villains
```

```python
# Ranking them
most_debuts = villains['Decade']
most_debuts.value_counts()
```

```python
# Assingment 3

companies = villains['Company']
companies.value_counts()
```
