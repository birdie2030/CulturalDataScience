```r
data(GSSvocab)
```

```r
str(GSSvocab)
```

```r
# filtering for the year 1978 and removing rows
data1978 <- GSSvocab %>%
    filter(year==1978) %>%
    na.exclude(data1978)

str(data1978)
```

```r
# Visualizing the relationsship between a person’s score on the vocabulary test and their level  of education
ggplot(data1978,
       aes(x = educ,
           y = vocab)) +
    geom_point(position = 'jitter')
```

```r
cor.test(data1978$vocab, data1978$educ)
```

```r
# Making a linear model of the correlation
vocab_model <- lm(vocab ~ educ, data1978)
summary(vocab_model)
```

```r
# There seems to be a  correlation between the two variables because the p-value seems to be signifcant.
# From the plot it shows that the correlation is positive.
# However the value of r-squared indicates that it is only about 29% of the variance that can be accounted for by the model.
# In conclusion, a person’s score on the vocabulary test is somewhat significantly impacted by their level of education
```

```r
# Visualizing the relationship between vocab, education and nativeborn
ggplot(data1978,
       aes(x = educ,
           y = vocab,
          color = nativeBorn)) +
    geom_point(position = 'jitter',
              alpha = 0.5) +
theme_minimal()
```

```r
vocab_model2 <- lm(vocab ~ educ + nativeBorn, data1978)
summary(vocab_model2)
```

```r
# The p-value is less than 0.05 so there is a relationship between nativeborn and vocab
```

```r
ggplot(data1978,
       aes(x = educ,
           y = nativeBorn)) +
geom_boxplot()
```

```r
# The median appears to be the same so nativeborn does not to seem to effect education
```

```r
vocab_model3 <- lm(vocab ~ educ * nativeBorn, data1978)
summary(vocab_model3)
```

```r
# the p-values are too high and therefore it does not make sense to add the interaction between nativeborn and education
```
